# Winners

```python
plt.figure(figsize=(34,8))
sns.countplot(df['winner'])
plt.show()
```

# Toss_Winners

```python
plt.figure(figsize=(34,8))
sns.countplot(df['toss_winner'])
plt.show()
```

# Toss_Winner Decision

```python
#plt.figure(figsize=(8,8))
sns.countplot(df['toss_decision'])
plt.show()
plt.show()
```

# Host_city

```python
plt.figure(figsize=(34,8))
sns.countplot(df['city'])
plt.show()
```

# venue

```python
plt.figure(figsize=(150,20))
sns.countplot(df['venue'])
plt.show()
```

# result

```python
#plt.figure(figsize=(20,20))
sns.countplot(df['result'])
plt.show()
```

# Top Man of the match Winners:

```python
plt.figure(figsize=(20,6))
sns.barplot(y=df['player_of_match'].value_counts().head(10).values,x=df['player_of_match'].value_counts().head(10).index)
plt.show()
```

# Top 10 Highest Match Venues:

```python
plt.pie(labels=df['city'].value_counts().head(10).index,x=df['city'].value_counts().head(10).values)
plt.show()
```

# Top 10 Umpire1:

```python
plt.title("Top umpire1:",loc='left')
plt.pie(labels=df['umpire1'].value_counts().head(10).index,x=df['umpire1'].value_counts().head(10).values)
plt.show()
```

```python
plt.title("Top umpire2:",loc='left')
plt.pie(labels=df['umpire2'].value_counts().head(10).index,x=df['umpire1'].value_counts().head(10).values)
plt.show()
```
